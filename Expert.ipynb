```python
from qiskit.quantum_info import Pauli
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

ZZ=Pauli('ZZ')
ZI=Pauli('ZI')
IZ=Pauli('IZ')
XX=Pauli('XX')
XI=Pauli('XI')
IX=Pauli('IX')

observables=[ZZ,ZI,IZ,XX,XI,IX]
```

```python
from qiskit_aer.primitives import Estimator
from qiskit_aer.utils import *

estimator=Estimator()
job=estimator.run([qc]*len(observables),observables)

result=job.result()
print(result)
```

```python
import matplotlib.pyplot as plt

data = ['ZZ', 'ZI', 'IZ', 'XX', 'XI', 'IX']
values = job.result().values

plt.bar(data, values)
plt.xlabel('Observables')
plt.ylabel('Expectation value')
plt.show()
```

$n$-qubit GHZ state

```python
def get_qc_for_n_qubit_GHZ_state(qc,n):
    qc.h(0)
    for i in range(n-1):
        qc.cx(i,i+1)
    return qc

n=100
qc = QuantumCircuit(n)
qc = get_qc_for_n_qubit_GHZ_state(qc,n)
qc.draw()

```
